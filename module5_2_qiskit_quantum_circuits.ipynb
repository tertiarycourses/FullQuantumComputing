## Adding Quantum Sub Circuits

```
from qiskit import  QuantumRegister, ClassicalRegister, QuantumCircuit
```

```
# Create a Quantum Register with 1 qubits and Classsical Register with 1 bit
q0 = QuantumRegister(1)
c0 = ClassicalRegister(1)
```

```
# Create a Quantum Circuit acting on the q and c registers
circ1 = QuantumCircuit(q0,c0)
```

```
# Add a X gate on qubit 0 to flip the bit
circ1.x(q0[0])
```

```
circ1.draw()
```

```
# Create a Quantum Register with 1 qubits and Classsical Register with 1 bit
q1 = QuantumRegister(1)
c1 = ClassicalRegister(1)
```

```
# Create a Quantum Circuit acting on the q and c registers
circ2 = QuantumCircuit(q1,c1)
```

```
# Add a Hadamad gate on qubit 0 to flip the bit
circ2.h(q1[0])
```

```
circ2.draw()
```

```
circ3 = circ1+circ2
```

```
circ3.draw()
```

## Measurement

```
# add measurement operation
circ3.measure(q0,c0)
circ3.measure(q1,c1)
```

```
#drawing the circuit
circ3.draw()
```

## Simulation

```
# Import Aer
from qiskit import BasicAer
from qiskit import execute                         
```

```
backend = BasicAer.get_backend('qasm_simulator')
job = execute(circ3, backend, shots=1024)
#job = qiskit.execute(circ, qiskit.BasicAer.get_backend('qasm_simulator') )

# Grab the results from the job.
result = job.result()
```

```
counts = result.get_counts()
print(counts)
```

```
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```

## Quantum Circuits Attributes

```
#total number of operations in the circuit.
circ3.size()
```

```
# depth of circuit (number of ops on the critical path)
circ3.depth()
```

```
# number of qubits in the circuit
circ3.width()
```

```
# a breakdown of operations by type
circ3.count_ops()
```

```
# number of unentangled subcircuits in this circuit.
# each subcircuit can in principle be executed on a different quantum processor!
circ3.num_tensor_factors()
```
